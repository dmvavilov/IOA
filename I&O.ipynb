```r
wiot2014 <- read.csv('Downloads/wiot14.csv', stringsAsFactors = FALSE)
```

```r
Z <- wiot2014[2:2465,3:2466]
row.names(Z) <- colnames(Z)
```

```r
VALU <- wiot2014[2471, 3:2466]
row.names(VALU) <- wiot2014[2471,1]
```

```r
X <-wiot2014[2473, 3:2466]
row.names(X) <- wiot2014[2473,1]
```

```r
FD <- wiot2014[2:2465,2467:2686]
row.names(FD) <- row.names(Z)
```

```r
Z <- as.matrix(apply(Z, 2, as.numeric))
VALU <- as.matrix(apply(VALU, 2, as.numeric))
X <- as.matrix(apply(X, 2, as.numeric))
FD <- as.matrix(apply(FD, 2, as.numeric))
```

```r
PROD <- rowSums(Z)
```

```r
ones <- matrix(data = 1, nrow = nrow(Z), ncol = ncol(Z))
for (i in 1:44) {
    ones[(i*56-55):(56*i), (i*56-55):(56*i)] <- 0
}
```

```r
ones1 <- matrix(data = 1, nrow = nrow(FD), ncol = ncol(FD))
for (i in 1:44) {
    ones1[(i*4-3):(4*i), (i*4-3):(4*i)] <- 0
}
```

```r
A <- Z/PROD
```

```r
A[is.na(A)] <- 0
is.na(A)<-sapply(A, is.infinite)
A[is.na(A)]<-0
```

```r
B <- solve(diag(nrow(Z)) - A)
```

```r
V <- VALU/PROD
V[is.na(V)] <- 0
is.na(V)<-sapply(V, is.infinite)
V[is.na(V)]<-0
```

```r
V_hat <- diag(as.vector(V))
```

##### Если мир рассматриваем как партнера,то нужно найти экспорт индустрии в каждую индустрию + FD (но за вычетом ситуации, когда происходит экспорт из страны в страну)

```r
EXGR_DVA <- rowSums(V_hat %*% B %*% diag(rowSums(cbind((Z*ones),(FD*ones1))))) 
```

```r
EXGR_DVA <- as.data.frame(EXGR_DVA)
```

```r
EXGR_DVA$Industry <- colnames(wiot2014[,3:2466])
```

```r
EXGR_DVA[which(EXGR_DVA$EXGR_DVA==max(EXGR_DVA$EXGR_DVA)),]
```

```r
EXGR_DVA <- EXGR_DVA[order(EXGR_DVA$EXGR_DVA, decreasing = T),]
```
